# Gerando gráficos - parte 2 (Análise descritiva)

Prática de análise descritiva de variáveis contínuas na base do DataSUS.

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('sinasc.csv')
df.head(3)
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Método 1 (Pandas)
df[['IDADEMAE']].boxplot()
```

```python
# Método 2 (via seaborn)
# Coordenada y
sns.boxplot(y='IDADEMAE', data=df)
```

```python
# coordenada x
sns.boxplot(x='IDADEMAE', data=df)
```

####  2) Faça um boxplot do peso do bebê

```python
# Via Pandas
df[['PESO']].boxplot()
```

```python
# Utilizando Seaborn
# eixo y
sns.boxplot(
    y='PESO', 
    data=df, 
    color='purple', 
    linecolor='black'
)
```

```python
# eixo x
sns.boxplot(x='PESO', 
            data=df, 
            color='blue', 
            linecolor='black'
)
```

####  3) Faça um histograma da idade da mãe

```python
# Método direto
fig = plt.figure(figsize=(4,4))
df['IDADEMAE'].plot.hist(bins=20)
```

```python
# Utilizando o Seaborn

sns.displot(
    data = df, 
    x = 'IDADEMAE',
    bins = 20,
    alpha = 0.6,
    # element = 'step',
    kde = True
)
```

#### 4) Faça um histograma do peso do bebê

```python
# Pode-se fazer
df['PESO'].plot.hist(bins=50)
```

```python
# Utilizando o seaborn

sns.displot(
    data = df,
    x = 'PESO',
    alpha = 0.7,
    bins = 50,
    element = 'bars',
    kde = True
)
```
